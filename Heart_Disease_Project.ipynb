```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import classification_report, confusion_matrix

# Define column names
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 
           'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

# Load dataset with correct column names
df = pd.read_csv('heart_disease/processed.cleveland.data', names=columns)

# Data Preprocessing
# Handle missing values - replace '?' with NaN and then fill with the mode for categorical columns and mean for numerical columns
df.replace('?', np.nan, inplace=True)
df['sex'] = df['sex'].fillna(df['sex'].mode()[0])
df['cp'] = df['cp'].fillna(df['cp'].mode()[0])
df['restecg'] = df['restecg'].fillna(df['restecg'].mode()[0])
df['slope'] = df['slope'].fillna(df['slope'].mode()[0])
df['ca'] = df['ca'].fillna(df['ca'].mode()[0])
df['thal'] = df['thal'].fillna(df['thal'].mode()[0])

# For numerical columns, fill NaN with mean
for col in ['age', 'trestbps', 'chol', 'fbs', 'thalach', 'exang', 'oldpeak']:
    df[col] = df[col].fillna(df[col].mean())

# Encoding categorical variables (converting to numeric)
df['sex'] = df['sex'].map({'Male': 1, 'Female': 0})
df['cp'] = df['cp'].map({'Typical Angina': 1, 'Atypical Angina': 2, 'Non-anginal Pain': 3, 'Asymptomatic': 4})
df['restecg'] = df['restecg'].map({'Normal': 0, 'Abnormality': 1})  # Assuming 2 categories here, adjust if needed
df['slope'] = df['slope'].map({'Up': 1, 'Flat': 0, 'Down': -1})  # Adjust according to your dataset
df['ca'] = df['ca'].map({0: 0, 1: 1, 2: 2, 3: 3, 4: 4})  # Assuming these are numerical
df['thal'] = df['thal'].map({'Normal': 1, 'Fixed Defect': 2, 'Reversable Defect': 3})

# Feature Scaling (Standardization)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('target', axis=1))

# Splitting the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(scaled_features, df['target'], test_size=0.2, random_state=42)

# Feature Selection using Recursive Feature Elimination (RFE)
model = RandomForestClassifier(n_estimators=100)

# Apply RFE
rfe = RFE(estimator=model, n_features_to_select=5)  # Select top 5 features
fit = rfe.fit(X_train, y_train)

# Get selected features
selected_features = df.drop('target', axis=1).columns[fit.support_]
print(f"Selected Features: {selected_features}")

# Train the model using selected features
X_train_selected = X_train[:, fit.support_]
X_test_selected = X_test[:, fit.support_]

model.fit(X_train_selected, y_train)

# Evaluate the model
y_pred = model.predict(X_test_selected)

# Print classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title("Confusion Matrix")
plt.show()

# Optional: Feature importance from RandomForestClassifier
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

# Visualizing Feature Importance
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train_selected.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train_selected.shape[1]), selected_features[indices], rotation=90)
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_imputed, y)

# Feature Scaling
scaler = StandardScaler()
X_rescaled = scaler.fit_transform(X_resampled)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_rescaled, y_resampled, test_size=0.2, random_state=42)

# Logistic Regression with balanced class weights
model = LogisticRegression(class_weight='balanced', solver='lbfgs', max_iter=1000)

# Train the model
model.fit(X_train, y_train)

# Predict the labels
y_pred = model.predict(X_test)

# Evaluate the model performance
print(f"Logistic Regression Performance:")
print(classification_report(y_test, y_pred))

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr')
print(f"ROC AUC Score: {roc_auc}")

# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy}")
```
